```python
import requests
from bs4 import BeautifulSoup
```

```python
election_data_dict = {"2024 Presidential": "https://results.vote.wa.gov/results/20241105/president-vice-president_bycounty.html",
                      "2024 Governor": "https://results.vote.wa.gov/results/20241105/governor_bycounty.html",
                      "2024 Secretary of State": "https://results.vote.wa.gov/results/20241105/secretary-of-state_bycounty.html",
                      "2024 Attorney General": "https://results.vote.wa.gov/results/20241105/attorney-general_bycounty.html",
                      }
```

```python
def scrape_from_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        return soup
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None
```

```python
url = "https://results.vote.wa.gov/results/20241105/president-vice-president_bycounty.html"

soup = scrape_from_url(url)

for line in soup.find_all():
    print(line)
```
